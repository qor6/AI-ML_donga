[5] 조건문

<비교연산자>

x < y : x가 y보다 작다

x > y : y가 x보다 크다

x == y : x와 y가 같다

x != y : x와 y가 같지 않다

x >= y : x가 y보다 크거나 같다

x <= y : x가 y보다 작거나 같다

<연산자>

x or y : x와 y 둘 중에 하나만 참이어도 참이다

x and y : x와 y 모두 참이어야 참이다

not x : x가 거짓이면 참이다

```python
grade = 91

if grade > 90:
  print("A 학점")
else:
  print("B ~ F 학점")
```

```python
grade = 75

if grade > 90:
  print("A 학점")
elif grade > 80:
  print("B 학점")
elif grade > 70:
  print("C 학점")
elif grade > 60:
  print("D 학점")
else:
  print("F 학점")
  
```

```python
money = 2000
card = True

if money > 3000 or card:
  print("택시")
else:
  print("도보")
```

```python
def add_function(a,b):
  c = a + b
  return c
```

```python
add_function(4,5)
```

```python
add_function("난","21")
```

```python
def list_summation(lst):
  result = 0
  for i in lst:
    result += i
  return result
```

```python
list_a = range(100)
list_summation(list_a)  # 0 ~ 99까지의 합
```

```python
print(list_a)
```

```python
def list_average(lst):
  total = 0
  for i in lst:
    total += i
  average = total / len(lst)
  return average
```

```python
list_a = range(100)
list_average(list_a) # 0 ~ 99까지의 합 나누기 개수(100:0~99)
```
