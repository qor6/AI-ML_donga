# Logic Gate of Perceptron
(AND, OR, NAND, NOR)

AND w = 0.5, b = -0.7

OR  w = 0.5, b = -0.2

NAND w = -0.5, b = 0.7 / if문 조건을 반대로

NOR 도 위의 NAND와 마찬가지로

## AND Gate

```python
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  s = np.sum(w*x) + b
  if s <= 0:
    return 0  #거짓
  else:
    return 1  #참
```

```python
print(AND(0,0))
print(AND(1,1))
print(AND(0,1))
print(AND(1,0))
```

## OR Gate


```python
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  s = np.sum(w*x) + b
  if s <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(1,1))
print(OR(0,1))
print(OR(1,0))
```

## NAND Gate

```python
print(NAND(0,0))
print(NAND(1,1))
print(NAND(0,1))
print(NAND(1,0))
```

## XOR Gate(NAND,OR,AND)


```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0,0))
print(XOR(1,1))
print(XOR(0,1))
print(XOR(1,0))
```
